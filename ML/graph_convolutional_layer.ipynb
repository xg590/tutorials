## Basics of graph convolutional layer 
See this directed graph, node 0 and 2 is neighbors of node 3 but node 3 is not (OK, node 1 is) the neighbor of node 0. 

<b>Note: feature of node 0 is [0, 0, 0], of node 1 is [1, 1, 1], of node 2 is [2, 2, 2] and node 3 is [3, 3, 3].</b>
<div>  </div>
if we multiply the adjacency matrix <b><i>A</i></b> of the graph by the feature matrix <b><i>X</i></b>, the new feature matrix is interesting.

```python
import numpy as np

A = np.matrix([
    [0, 1, 0, 0],
    [0, 0, 1, 1], 
    [0, 1, 0, 0],
    [1, 0, 1, 0]],
    dtype=float
)

X = np.matrix([[i, i, i] for i in range(A.shape[0])], dtype=float)
_H = A * X 
_H
```

Node 0 and 2 is neighbors of node 3 so the feature of node 3 is [0, 0, 0]+[2, 2, 2] after the multiplication. This is how the graph is updated.
<div>  </div>
Graph convolutional layer: By mutiplying <b><i>AX</i></b> with weight matrix <b><i>W</i></b> plus bias <b><i>b</i></b>, we achieved a step of forward propagation.

```python
W = np.matrix([[0, 0], [1, 1], [2, 2]])
H = _H * W
H
```
