```python
import matplotlib.pyplot as plt, numpy as np
from matplotlib import font_manager 

old_font = set([f.name for f in font_manager.fontManager.ttflist])

font_files = font_manager.findSystemFonts(fontpaths=['/tmp/font/'])
[print(i) for i in sorted(font_files)]

[font_manager.fontManager.addfont(font_file) for font_file in font_files]    

new_font = set([f.name for f in font_manager.fontManager.ttflist])

new_font.difference(old_font)
```

```python
# ---------------------------------------------------
fig = plt.figure(figsize=(8,8), layout="constrained")
gs  = fig.add_gridspec(2, 2)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1]) 
ax3 = fig.add_subplot(gs[1, 0:2])
  
t1 = np.arange(0.0, 3.0, 0.01)
f = lambda t: np.exp(-t) * np.cos(2*np.pi*t)  
ax1.plot(t1, f(t1)) 
ax2.plot(t1, f(t1))  
ax3.plot(t1, f(t1))  


ax1.set_title("Arial", fontdict={'fontsize': 72, 'fontweight': 'medium','family':'Arial'}) 
ax2.set_title("Courier New", fontdict={'fontsize': 22, 'fontweight': 'medium','family':'Courier New'}) 
ax3.set_title("Time New Roman", fontdict={'fontsize': 32, 'fontweight': 'medium','family':"Times New Roman"}) 
 

plt.show() 
```
